```python
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math

with open('const_delta_fidelity_h_1q.pkl', 'rb') as f:
    F_h = pickle.load(f)
f.close()

with open('const_delta_fidelity_ch_1q.pkl', 'rb') as f:
    F_ch = pickle.load(f)
f.close()

with open('const_delta_list_1q.pkl', 'rb') as f:
    delta_list = pickle.load(f)
f.close()

with open('thr_const_delta_f_hamiltonian_1q.pkl', 'rb') as f:
    F_h_thr = pickle.load(f)
f.close()

with open('thr_const_delta_f_channel_1q.pkl', 'rb') as f:
    F_ch_thr = pickle.load(f)
f.close()

with open('thr_const_delta_list_1q.pkl', 'rb') as f:
    delta_list_thr = pickle.load(f)
f.close()

F_h_thr = [x * 100 for x in F_h_thr]
F_ch_thr = [x * 100 for x in F_ch_thr]
print(F_h)
```

```python
plt.figure(dpi=1200)
plt.title('F_ave with constant dephasing noise')

plt.plot(delta_list, F_ch, 'ro', markersize=2, label='Channel noise simulation')
plt.plot(delta_list, F_h, 'bo', markersize=2, label='Hamiltonian noise simulation')

plt.plot(delta_list_thr, F_ch_thr, '--', color='tab:red', linewidth=1, label='Channel noise theoetical fidelity')
plt.plot(delta_list_thr, F_h_thr, '--', color='tab:blue', linewidth=1, label='Hamiltonian noise theoetical fidelity')

plt.xlabel("Noise strength delta (rad)")
plt.ylabel("Clifford average gate fidelity (%)")
plt.legend()
plt.show()
```
