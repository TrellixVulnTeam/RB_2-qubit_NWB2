```python
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import numpy as np

with open('const_delta_fidelity_h_1q_500.pkl', 'rb') as f:
    F_h_500 = pickle.load(f)
f.close()

with open('const_delta_list_h_1q_500.pkl', 'rb') as f:
    delta_list_h_500 = pickle.load(f)
f.close()

with open('const_delta_fidelity_ch_1q_500.pkl', 'rb') as f:
    F_ch_500 = pickle.load(f)
f.close()

with open('const_delta_list_ch_1q_500.pkl', 'rb') as f:
    delta_list_ch_500 = pickle.load(f)
f.close()

with open('const_delta_fidelity_h_1q_2000.pkl', 'rb') as f:
    F_h_2000 = pickle.load(f)
f.close()

with open('const_delta_list_h_1q_2000.pkl', 'rb') as f:
    delta_list_h_2000 = pickle.load(f)
f.close()

with open('const_delta_fidelity_h_1q_2000_rdini.pkl', 'rb') as f:
    F_h_2000_rdini = pickle.load(f)
f.close()

with open('const_delta_list_h_1q_2000_rdini.pkl', 'rb') as f:
    delta_list_h_2000_rdini = pickle.load(f)
f.close()

with open('thr_const_delta_f_hamiltonian_avg_1q.pkl', 'rb') as f:
    F_h_avg = pickle.load(f)
f.close()

with open('thr_const_delta_f_hamiltonian_rb_1q.pkl', 'rb') as f:
    F_h_rb = pickle.load(f)
f.close()

with open('thr_const_delta_f_channel_1q.pkl', 'rb') as f:
    F_ch_thr = pickle.load(f)
f.close()

with open('thr_const_delta_list_1q.pkl', 'rb') as f:
    delta_list_thr = pickle.load(f)
f.close()

with open('ensemble_sigma_fidelity_h_1q_500_fastest.pkl', 'rb') as f:
    F_h_ens_rb = pickle.load(f)
f.close()

with open('ensemble_sigma_fidelity_h_1q_500_slowest.pkl', 'rb') as f:
    F_h_ens_rb_2 = pickle.load(f)
f.close()

with open('ensemble_sigma_list_1q_h_1q_500.pkl', 'rb') as f:
    sigma_list_ens = pickle.load(f)
f.close()

F_h_avg = [x * 100 for x in F_h_avg]
F_h_rb = [x * 100 for x in F_h_rb]
F_ch_thr = [x * 100 for x in F_ch_thr]
```

```python
plt.figure(dpi=1200)
plt.title('F_ave with constant dephasing noise')

delta_list = [x / (np.pi) for x in delta_list_h_500]
sigma_list = [x / (np.pi) for x in sigma_list_ens]
plt.plot(delta_list, F_h_500, 'ro', markersize=1, label='RB simulation')

plt.plot(delta_list, F_h_avg, '--', color='tab:blue', linewidth=1, label='Hamiltonian noise average fidelity', zorder=2)
plt.plot(delta_list, F_h_rb, '--', color='black', linewidth=1, label='Hamiltonian noise RB theoretical fidelity', zorder=1)

plt.xlabel("Noise strength delta")
plt.ylabel("Clifford average gate fidelity (%)")
plt.legend(prop={'size': 6})
plt.show()
```

```python
with open('thr_ensemble_sigma_f_hamiltonian_rb_1q.pkl', 'rb') as f:
    F_ens_h_thr = pickle.load(f)
f.close()

with open('thr_ensemble_sigma_f_hamiltonian_avg_1q.pkl', 'rb') as f:
    F_ens_h_avg = pickle.load(f)
f.close()

F_ens_h_avg = [x * 100 for x in F_ens_h_avg]
F_ens_h_thr = [x * 100 for x in F_ens_h_thr]
```

```python
plt.figure(dpi=1200)
plt.title('F_ave with ensemble dephasing noise')

delta_list = [x / (np.pi) for x in delta_list_h_500]
sigma_list = [x / (np.pi) for x in sigma_list_ens]

plt.plot(sigma_list, F_h_ens_rb, 'bo', markersize=1, label='fastest ensemble dephasing noise', zorder=3)
plt.plot(sigma_list, F_h_ens_rb_2, 'go', markersize=1, label='slowest ensemble dephasing noise', zorder=3)

plt.plot(delta_list, F_ens_h_avg, '--', color='tab:blue', linewidth=1, label='Hamiltonian noise average fidelity', zorder=2)
plt.plot(delta_list, F_ens_h_thr, '--', color='black', linewidth=1, label='Hamiltonian noise RB theoretical fidelity', zorder=1)

plt.xlabel("Noise strength sigma")
plt.ylabel("Clifford average gate fidelity (%)")
plt.legend(prop={'size': 6})
plt.show()
```

```python
with open('const_delta_r_squared_1q_constant.pkl', 'rb') as f:
    r_constant = pickle.load(f)
f.close()

with open('ensemble_sigma_r_squared_1q_fastest.pkl', 'rb') as f:
    r_fastest = pickle.load(f)
f.close()

with open('ensemble_sigma_r_squared_1q_slowest.pkl', 'rb') as f:
    r_slowest = pickle.load(f)
f.close()


plt.figure(dpi=1200)
plt.title('SRB Fitting R-squared comparison')

plt.plot(delta_list, r_constant, 'ro', markersize=1, label='constant dephasing noise')
plt.plot(sigma_list, r_fastest, 'bo', markersize=1, label='fastest ensemble noise')
plt.plot(sigma_list, r_slowest, 'go', markersize=1, label='slowest ensemble noise')


plt.xlabel("Noise strength delta/sigma for const/ensemble noise")
plt.ylabel("R-squared")
plt.legend(prop={'size': 6})
plt.show()
```

```python
with open('const_delta_pulse_tr_fidelity_h_1q.pkl', 'rb') as f:
    F_tr_x_pi2 = pickle.load(f)
f.close()

F_h_pulse_thr = [(x+100)/2 for x in F_h_avg]
F_h_pulse_2000_rdini = [(x+100)/2 for x in F_h_2000_rdini]
F_tr_x_pi2 = [x*100 for x in F_tr_x_pi2]
delta_list_tr = [x/2 for x in delta_list_h_2000]

plt.figure(dpi=1200)
plt.title('Trace fidelity and average pulse fidelity comparison')

plt.plot(delta_list_tr, F_tr_x_pi2, 'go', markersize=1, label='Trace fidelity of X_pi/2 pulse')
plt.plot(delta_list_tr, F_h_pulse_2000_rdini, 'bo', markersize=1, label='Average pulse fidelity')

plt.plot(delta_list_tr, F_h_pulse_thr, '--', color='tab:blue', linewidth=1, label='Theoretical average pulse fidelity')

plt.xlabel("Noise strength delta (rad)")
plt.ylabel("Fidelity (%)")
plt.legend(prop={'size': 6})
plt.show()
```
