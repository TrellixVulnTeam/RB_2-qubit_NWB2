```python
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import numpy as np
```

```python
with open('noise_switch_frequency.pkl', 'rb') as f:
    switch_list = pickle.load(f)
f.close()

switch_list_0 = np.concatenate((switch_list[:10],switch_list[:10], switch_list[:10]), axis=None)
switch_list_dt = np.concatenate((switch_list[10:],switch_list[10:], switch_list[10:]), axis=None)
```

```python
with open('F_Clifford 1.pkl', 'rb') as f:
    F_Clifford_1 = pickle.load(f)
f.close()
with open('r_sqrd 1.pkl', 'rb') as f:
    r_sqrd_1 = pickle.load(f)
f.close()
with open('F_Clifford 2.pkl', 'rb') as f:
    F_Clifford_2 = pickle.load(f)
f.close()
with open('r_sqrd 2.pkl', 'rb') as f:
    r_sqrd_2 = pickle.load(f)
f.close()
with open('F_Clifford 3.pkl', 'rb') as f:
    F_Clifford_3 = pickle.load(f)
f.close()
with open('r_sqrd 3.pkl', 'rb') as f:
    r_sqrd_3 = pickle.load(f)
f.close()
# data set 4 has 5x noise
with open('F_Clifford 4.pkl', 'rb') as f:
    F_Clifford_4 = pickle.load(f)
f.close()
with open('r_sqrd 4.pkl', 'rb') as f:
    r_sqrd_4 = pickle.load(f)
f.close()

F_Clifford_0 = np.concatenate((F_Clifford_1[:10], F_Clifford_2[:10], F_Clifford_3[:10]), axis=None) * 100
F_Clifford_dt = np.concatenate((F_Clifford_1[10:], F_Clifford_2[10:], F_Clifford_3[10:]), axis=None) * 100
r_sqrd_0 = np.concatenate((r_sqrd_1[:10], r_sqrd_2[:10], r_sqrd_3[:10]), axis=None)
r_sqrd_dt = np.concatenate((r_sqrd_1[10:], r_sqrd_2[10:], r_sqrd_3[10:]), axis=None)

F_Clifford = np.array([np.mean(F_Clifford_0), np.mean(F_Clifford_dt)])
F_err = np.array([np.std(F_Clifford_0), np.std(F_Clifford_dt)])
r_sqrd = np.array([np.mean(r_sqrd_0), np.mean(r_sqrd_dt)])
r_err = np.array([np.std(r_sqrd_0), np.std(r_sqrd_dt)])

print(F_Clifford_4)
```

```python
xtick = [0, 2e2]

plot1 = plt.figure(1)
plt.plot(switch_list_0/1000000, F_Clifford_0, 'o', markersize=2)
plt.plot(switch_list_dt/1000000, F_Clifford_dt, 'o', markersize=2)
plt.xticks(xtick)
# plt.xlabel("Switching time (us)")
plt.xlabel("Switching frequency (MHz)")
plt.ylabel("Clifford infidelity (%)")
plt.show()

plot2 = plt.figure(2)
plt.plot(switch_list_0/1000000, r_sqrd_0, 'o', markersize=2)
plt.plot(switch_list_dt/1000000, r_sqrd_dt, 'o', markersize=2)
plt.xticks(xtick)
# plt.xlabel("Switching time (us)")
plt.xlabel("Switching frequency (MHz)")
plt.ylabel("R_squared")
plt.show()

plot3 = plt.figure(3)
plt.errorbar(xtick, F_Clifford, F_err, fmt='go', markersize=2, capsize=3)
plt.xlabel("Switching frequency (Hz)")
plt.ylabel("Clifford infidelity (%)")
plt.show()

plot4 = plt.figure(4)
plt.errorbar(xtick, r_sqrd, r_err, fmt='go', markersize=2, capsize=3)
plt.xlabel("Switching frequency (Hz)")
plt.ylabel("R_squared")
plt.show()
```
