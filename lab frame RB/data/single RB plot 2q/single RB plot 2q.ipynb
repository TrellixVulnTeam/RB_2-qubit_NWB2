```python
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import numpy as np
```

```python
with open('16000_2q.pkl', 'rb') as f:
    data = pickle.load(f)
f.close()

F = data[1]
###############
# SRB with fixed B

def func(x, A, B, r):
    return A * (1 - 4 / 3 * r) ** x + B

l1 = np.arange(1, 20, 2)
l2 = np.arange(20, 100, 10)
L = np.hstack((l1, l2))
popt, pcov = curve_fit(func, L, F, p0=[1, 0, 0], bounds=(0, 1), maxfev=5000)
print("F_Ciff = 1 - r = ", (1 - popt[2]) * 100)

plt.plot(L, F, 'o', markersize=4)
plt.plot(L, func(L, *popt), 'r-')
plt.ylim(top=1.0)
plt.xlabel("Number of Cliffords (L)")
plt.ylabel("Proj. State Prob.")
plt.show()
```

```python
with open('50000_2q.pkl', 'rb') as f:
    data = pickle.load(f)
f.close()

F = data[1]
###############
# SRB with fixed B

def func(x, b, a):
    return 3/4 * 2.71828 ** (-(b*x)**a) + 1/4

l1 = np.arange(1, 20, 2)
l2 = np.arange(20, 100, 10)
L = np.hstack((l1, l2))
popt, pcov = curve_fit(func, L, F, p0=[1, 1], bounds=(0, 1), maxfev=5000)
print(popt)
print("F_Ciff = ", (1 - popt[0]) * 100)

plt.plot(L, F, 'o', markersize=4)
plt.plot(L, func(L, *popt), 'r-')
plt.ylim(top=1.0)
plt.xlabel("Number of Cliffords (L)")
plt.ylabel("Proj. State Prob.")
plt.show()
```
