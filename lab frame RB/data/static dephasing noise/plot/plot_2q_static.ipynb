```python
import pickle
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import numpy as np

with open('static_dephasing_noise_fidelity.pkl', 'rb') as f:
    F_static = pickle.load(f)
f.close()

with open('static_dephasing_noise_delta.pkl', 'rb') as f:
    delta_list = pickle.load(f)
f.close()

with open('thr_const_delta_f_hamiltonian_avg_2q.pkl', 'rb') as f:
    F_avg_thr = pickle.load(f)
f.close()

with open('thr_const_delta_f_hamiltonian_rb_2q.pkl', 'rb') as f:
    F_wallman_thr = pickle.load(f)
f.close()

with open('thr_const_delta_list_2q.pkl', 'rb') as f:
    delta_list_thr = pickle.load(f)
f.close()

F_avg_thr = [x * 100 for x in F_avg_thr]
F_wallman_thr = [x * 100 for x in F_wallman_thr]
```

```python
plt.figure(dpi=1200)
plt.title('2-qubit F_ave with constant dephasing noise')

plt.plot(delta_list, F_static, 'ro', markersize=1, label='Hamiltonian noise simulation')

plt.plot(delta_list_thr, F_avg_thr, '--', color='tab:blue', linewidth=1, label='Hamiltonian noise average fidelity', zorder=2)
plt.plot(delta_list_thr, F_wallman_thr, '--', color='black', linewidth=1, label='Hamiltonian noise RB theoretical fidelity', zorder=1)

plt.xlabel("Noise strength delta (Hz)")
plt.ylabel("Clifford average gate fidelity (%)")
plt.legend(prop={'size': 6})
plt.show()
```
